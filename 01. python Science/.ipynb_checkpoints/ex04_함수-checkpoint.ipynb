```python
def divisor(num) :
    for i in range(1,num):
        if num%i == 0:
            print(i, end= " ")
```

```python
divisor(100)
```

```python
def print_map(**kwargs):
    print(kwargs)
    for k, v in kwargs.items():
        print(k,":",v)
```

```python
#    
import random
lotto = []
while True:
    num = random.randint(1,45)
    if num not in lotto:
        lotto.append(num)
    if len(lotto) == 6:
        break;

lotto
```

```python
#   
inputStr = input("    6 ").split(" ")
mine = []
for n in inputStr:
    mine.append(int(n))
```

```python
#   lotto    
same = 0
for myN in mine:
    if myN in lotto:
        same += 1
```

```python
# same    
score = [0,6,5,3,2,1]
if same == 0:
    print('')
else :
    print(f'{same} , {score[same]} !')
```

```python
import numpy as np
def make_lotto(count):
    for i in range(count):
        lotto_num = [] #     

        for j in range(6): # 6
            lotto_num = np.random.choice(range(1,46), 6, replace=False)
        lotto_num.sort() #     
        print(f"{i+1}.  : {lotto_num}")
```

```python
count = int(input("    ? >> "))
make_lotto(count)
```
